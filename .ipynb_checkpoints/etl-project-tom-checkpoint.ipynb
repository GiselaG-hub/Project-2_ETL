```python
# Dependencies and Setup
import pandas as pd
import numpy as np
```

```python
# File to Load (Remember to Change These)
gun_violence_file = "Resources/gun-violence-data_01-2013_03-2018.csv"
full_moon_file = "Resources/full_moon.csv"
```

```python
# Read 1st file and store into Pandas data frame
gun_violence_data_df = pd.read_csv(gun_violence_file)
gun_violence_data_df.head()
gun_violence_data_df.shape
```

```python
# Read 2nd file and store into Pandas data frame
full_moon_data_df = pd.read_csv(full_moon_file)
full_moon_data_df.head()
```

```python
#rename DATE column in full_moon_data_df to match column in gun_violence_data_df
full_moon_data_df.rename(columns = {'DATE':'date'}, inplace = True) 
full_moon_data_df.head()
```

```python
full_moon_data_df['date'] = pd.to_datetime(full_moon_data_df['date'])
full_moon_data_df.head()
```

```python
gun_violence_data_df.dtypes
gun_violence_data_df['date']=pd.to_datetime(gun_violence_data_df['date'])
gun_violence_data_df.dtypes
```

```python
full_moon_data_df['date']=pd.to_datetime(full_moon_data_df['date'])
full_moon_data_df.dtypes
```

```python
#join the data into a third dataframe
merged_df=gun_violence_data_df.merge(full_moon_data_df, on='date',how = "left")

merged_df.head(200)
```

```python
crime_data_cols = ["incident_id","date", "state", "n_killed", "n_injured", "participant_relationship", "participant_type"]
merged_df.dropna(how="any")
merged_df.head()
```

```python
crime_data_cols = ["date", "state", "n_killed", "n_injured", "incident_characteristics", "Year"]
crime_data_transformed_df = merged_df[crime_data_cols].copy()
# Rename the column headers
crime_data_transformed_df = crime_data_transformed_df.rename(columns={"date": "Date",
                                                              "state": "State",
                                                              "n_killed": "Killed",
                                                              "n_injured": "Injured",
                                                              "incident_characteristics": "Incident Char",
                                                              "participant_type": "Type",
                                                                     "Year": "Full Moon"})
crime_data_transformed_df.head()
```
